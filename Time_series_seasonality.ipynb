```python
import numpy as np
from pandas import to_datetime as dt
import matplotlib.pyplot as plt
import pandas as pd
data={"Date": dt(["2001-01-03", "2002-02-20", "2003-03-10", "2004-04-05", "2005-05-26", "2006-06-24","2007-07-16", "2008-08-09", "2009-09-05", "2010-10-21", "2011-11-30", "2012-12-21"]),
      "Value": [10,7,25,38,65,47,56,86,58,67,36,90]}
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Value"], marker='o')
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Time Series with trend")
plt.grid(True)
plt.show()
```
